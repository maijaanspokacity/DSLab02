```python
import pandas as pd

df = pd.read_csv('accord_sedan.csv')

import matplotlib.pyplot as plt
import seaborn as sns

# Distribution plots
sns.histplot(df['price'], bins=50)
plt.title('Price Distribution')
plt.show()
```

```python
sns.histplot(df['mileage'], bins=50)
plt.title('Mileage Distribution')
plt.show()
```

```python
# Joint distribution
sns.scatterplot(x='mileage', y='price', data=df)
plt.title('Mileage vs. Price')
plt.show()
```

```python
def is_outlier(data):
    mean = data.mean()
    std = data.std()
    return (data < mean - 2 * std) | (data > mean + 2 * std)

df['isOutlierPrice'] = is_outlier(df['price'])
df['isOutlierMileage'] = is_outlier(df['mileage'])

sns.scatterplot(x='mileage', y='price', data=df, hue='isOutlierPrice')
plt.title('Mileage vs. Price with Outliers')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm


df = pd.read_csv('TB_burden_countries_2014-09-29.csv', header=0)
data = df['e_mort_exc_tbhiv_100k']

# Histogram
sns.histplot(data, kde=True)
plt.title('Histogram of e_mort_exc_tbhiv_100k')
plt.show()

# Q-Q plot
sm.qqplot(data, line='s')
plt.title('Q-Q Plot of e_mort_exc_tbhiv_100k')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

std_devs = [0.5, 1, 2]
sample_sizes = [5, 10, 100]

# Calculate statistics
def calculate_stats(data):
    mean = np.mean(data)
    std = np.std(data)
    kurt = kurtosis(data)
    return mean, std, skew, kurt

results = []
for std_dev in std_devs:
    for sample_size in sample_sizes:
        data = np.random.normal(loc=0, scale=std_dev, size=sample_size)
        mean, std, skew, kurt = calculate_stats(data)
        results.append([std_dev, sample_size, mean, std, skew, kurt])

import pandas as pd

results_df = pd.DataFrame(results, columns=["Std Dev", "Sample Size", "Mean", "Std Dev", "Skewness", "Kurtosis"])
print(results_df.to_markdown(index=False))


```
