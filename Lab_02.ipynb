```python
import pandas as pd
```

```python
passenger_data = pd.read_csv('passengerData.csv')
ticket_price = pd.read_excel('ticketPrices.xlsx')

data = passenger_data.merge(ticket_price, on='TicketType')
```

# Exercise 1

```python
data = data[data['Age'].notna()]

oldest_passenger_age = data['Age'].max()
oldest_passenger = data[data['Age'] == oldest_passenger_age]
print(f'The name of the oldest passenger is "{oldest_passenger["Name"].values[0]}"')
```

```python
data.plot.scatter('Age', 'Fare')
```

```python
female_40_50 = data[(data['Age'] >= 40) & (data['Age'] <= 50) & (data['Fare'] >= 40) & (data['Sex'] == 'female')]
female_40_50.plot.scatter('Age', 'Fare')
```
